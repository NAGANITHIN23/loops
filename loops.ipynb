```python
list1 = [1,22,33,44,55,66,77,88,99,1000]
for num in list1:
    print(num)
```

```python
for num in list1:
    if num % 11 == 0:
        print(num)
```

```python
for num in list1:
    if num % 11 == 0:
        print(num)
    else:
        print('not divisible')
```

```python
list_sum = 0 
for num in list1:
    list_sum = list_sum + num
print(list_sum)
```
